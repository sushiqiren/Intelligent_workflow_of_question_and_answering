# Set up pipeline for Llama2 inference

```python
from transformers import pipeline
llama_model = "hzsushiqiren/llama-2-7b-miniQuA"

generator = pipeline("text-generation", model = llama_model, max_length=200)
```

# Output answer with two-sentences format

```python
@anvil.server.callable
def get_answer(question):
    result = generator(question)
    list = result[0]['generated_text'].split("\n")
    answerSentences = " " . join(list[1:])
    answerSentencesList = answerSentences.split(".")
    answer = ""
    index = answerSentencesList[1].find("?")
    if index != -1:
        answer = answerSentencesList[0] + "." + answerSentencesList[1][(index + 1):] + "."
    else:
        answerString = "." . join(answerSentencesList[0:2])
        answer = answerString[1:] + "."

    return answer
```
